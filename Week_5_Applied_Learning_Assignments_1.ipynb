```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Find numerical columns with missing values
missing_before = df[numerical_cols].isnull().sum()
affected_cols = missing_before[missing_before > 0]

# Fill missing values with the mean and report
if affected_cols.empty:
    print(" No numerical columns have missing values.")
else:
    print(" Filling missing values with column means:\n")
    for col, count in affected_cols.items():
        mean_val = df[col].mean()
        df[col].fillna(mean_val, inplace=True)
        print(f"â€¢ {col}: {count} missing values filled")
```

```python
# Record original dataset size
original_rows = df.shape[0]

# Set threshold for missing values (50% of columns)
threshold = df.shape[1] * 0.5

# Identify rows to be dropped
rows_to_drop = df[df.isnull().sum(axis=1) > threshold]

# Capture enrollee_id of dropped rows
dropped_ids = rows_to_drop['enrollee_id'].tolist()

# Drop the rows
df_cleaned = df.dropna(thresh=threshold)

# Report impact
new_rows = df_cleaned.shape[0]
dropped_rows = original_rows - new_rows

# Print summary
print(f" Dropped {dropped_rows} rows where more than 50% of data was missing.")
print(f" Dataset reduced from {original_rows} to {new_rows} rows.")
print(" Enrollee IDs of dropped rows:")
print(dropped_ids)
```

```python
# Parse experience using math logic
def parse_experience(val):
    if pd.isnull(val):
        return None
    val = str(val).strip()
    if val.startswith('<'):
        return float(val[1:]) - 0.1
    if val.startswith('>'):
        return float(val[1:]) + 0.1
    if '+' in val:
        return float(val.replace('+', '')) + 0.1
    if '-' in val:
        try:
            return float(val.split('-')[0])
        except:
            return None
    try:
        return float(val)
    except:
        return None

# Parse company size using math logic
def parse_company_size(val):
    if pd.isnull(val):
        return None
    val = str(val).strip()
    if re.match(r'^\d+\+$', val):
        return float(val.replace('+', '')) + 0.1
    elif re.match(r'^\d+-\d+$', val):
        try:
            return float(val.split('-')[0])
        except:
            return None
    else:
        return None  # Ignore malformed entries

# Apply parsing functions
df['experience_num'] = df['experience'].apply(parse_experience)
df['company_size_min'] = df['company_size'].apply(parse_company_size)

# Filter rows: Experience > 10 and Company_Size == 7 (i.e. >10000 employees)
filtered_df = df[
    (df['experience_num'] > 10) &
    (df['company_size_min'] > 7)
]

# Display results
print(filtered_df)
```

```python
# Parse 'experience' into numeric values
def parse_experience(val):
    if pd.isnull(val):
        return None
    val = str(val).strip()
    if val.startswith('<'):
        return float(val[1:]) - 0.1
    if val.startswith('>'):
        return float(val[1:]) + 0.1
    if '+' in val:
        return float(val.replace('+', '')) + 0.1
    if '-' in val:
        try:
            return float(val.split('-')[0])
        except:
            return None
    try:
        return float(val)
    except:
        return None

# Parse 'last_new_job' into numeric values
def parse_last_new_job(val):
    if pd.isnull(val):
        return None
    val = str(val).strip()
    if val == 'never':
        return 0
    if val.startswith('<'):
        return float(val[1:]) - 0.1
    if val.startswith('>'):
        return float(val[1:]) + 0.1
    try:
        return float(val)
    except:
        return None

# Apply parsing functions
df['experience_num'] = df['experience'].apply(parse_experience)
df['last_new_job_num'] = df['last_new_job'].apply(parse_last_new_job)

# Create the new feature
df['experience_gap'] = df['experience_num'] - df['last_new_job_num']

# Display the updated DataFrame for the first 20 only
print(df[['experience', 'last_new_job', 'experience_gap']].head(6))
```

```python
# Min-max normalization of city_development_index
df['city_development_index_normalized'] = (
    df['city_development_index'] - df['city_development_index'].min()
) / (
    df['city_development_index'].max() - df['city_development_index'].min()
)

# Display the first 10 rows of original and normalized city development index
print(" First 10 rows showing original and normalized city development index:")
print(df[['city_development_index', 'city_development_index_normalized']].head(10))
```

```python
# Display the top 10 cities with the highest normalized development index
df_sorted = df.sort_values(by='city_development_index_normalized', ascending=False)
print("\nðŸ”¹ Top 10 cities with the highest normalized development index:")
print(df_sorted[['city_development_index', 'city_development_index_normalized']].head(10))
```

###  ***Why Normalization Matters***

**1. Equal feature influence:**  
Prevents features with large numeric ranges from dominating smaller-scaled ones in models like logistic regression or neural networks.

**2. Improved accuracy:**  
Algorithms that rely on distance such as k-Nearest Neighbors (k-NN) and k-Means clustering perform more effectively when all features are on a comparable scale.

**3. Faster training:**  
Optimization methods like gradient descent converge more efficiently when input features are normalized.
